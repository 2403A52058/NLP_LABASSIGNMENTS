```python
# Import libraries
import nltk
import spacy

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Download NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Medical text corpus
medical_text = """
The patient was diagnosed with diabetes mellitus.
Prescribed medications include insulin injections.
Doctors are monitoring blood glucose levels regularly.
Early diagnosis helps in preventing complications.
"""
```

```python
import nltk

# Download required NLTK tokenizers
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Sentence tokenization using NLTK
sentences = sent_tokenize(medical_text)

print("Sentences:")
for s in sentences:
    print("-", s)
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(medical_text)

print("Sentences:")
for s in sentences:
    print("-", s)
```

```python
# Import word tokenizer
from nltk.tokenize import word_tokenize

# Perform word tokenization
words = word_tokenize(medical_text)

# Display words
print("Words:")
print(words)
```

```python
# Load spaCy English model
import spacy
nlp = spacy.load("en_core_web_sm")

# Process medical text
doc = nlp(medical_text)

# Display spaCy tokens
print("spaCy Tokens:")
for token in doc:
    print(token.text)
```

```python
# Import stemmer
from nltk.stem import PorterStemmer

# Initialize stemmer
stemmer = PorterStemmer()

# Apply stemming
stemmed_words = [stemmer.stem(word) for word in words]

# Display stemmed words
print("Stemmed Words:")
print(stemmed_words)
```

```python
# Import lemmatizer
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

# Display lemmatized words
print("Lemmatized Words:")
print(lemmatized_words)
```

```python
# Display spaCy lemmatization
print("spaCy Lemmatization:")
for token in doc:
    print(f"{token.text} → {token.lemma_}")
```

```python
# Compare stemming and lemmatization side by side

print(f"{'Word':<15}{'Stemmed':<15}{'Lemmatized':<15}")
print("-" * 45)

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:<15}{stem:<15}{lemma:<15}")
```
